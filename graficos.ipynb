```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

meses = ['jan', 'fev', 'mar', 'abr', 'mai', 'jun', 'jul', 'ago', 'set', 'out', 'nov', 'dez']
dados = {
    'alta': [26, 25, 24, 22, 19, 17, 17, 19, 20, 22, 24, 25],
    'media': [20, 20, 19, 16, 13, 11, 10, 12, 14, 16, 18, 20],
    'baixa': [15, 15, 14, 12, 8, 7, 6, 7, 9, 11, 13, 15],
}

ax.set_title('Média de temperatura nos últimos 12 meses em Lages')
ax.set_xlabel('Mês')
ax.set_ylabel('Temperatura')

ax.plot(meses, dados['alta'], label='Alta')
ax.plot(meses, dados['media'], label='Média')
ax.plot(meses, dados['baixa'], label='Baixa')
ax.legend()

plt.show()
```

```python
import cmath

def f(x):
    return cmath.exp(x)

dadosx = np.arange(-100, 100)
dadosy = [f(x) for x in dadosx]

plt.plot(dadosx, dadosy)
plt.show()
```
